#### what is hypothesis Testing? 
A statistical method that is used in making statistical decisions using experimental data. 
An assumption that we making about a population parameter. 

        # example: Avg student in class is 40 or a boy is taller than girls 
        

```python
# Packages
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# define constants
mu = 998.8 
sigma = 73.10
x1 = 900
x2 = 1100
```

```python
# calculate the z-transform
z1 = ( x1 - mu ) / sigma
z2 = ( x2 - mu ) / sigma
```

```python
x = np.arange(z1, z2, 0.001) # range of x in spec
x_all = np.arange(-10, 10, 0.001) # entire range of x, both in and out of spec
# mean = 0, stddev = 1, since Z-transform was calculated
y = norm.pdf(x,0,1)
y2 = norm.pdf(x_all,0,1)
```

```python
# build the plot
fig, ax = plt.subplots(figsize=(9,6))
plt.style.use('fivethirtyeight')
ax.plot(x_all,y2)

ax.fill_between(x,y,0, alpha=0.3, color='b')
ax.fill_between(x_all,y2,0, alpha=0.1)
ax.set_xlim([-4,4])
ax.set_xlabel('# of Standard Deviations Outside the Mean')
ax.set_yticklabels([])
ax.set_title('Normal Gaussian Curve')

plt.savefig('normal_curve.png', dpi=72, bbox_inches='tight')
plt.show()
```

```python
# Another Normal Distribution
import numpy as np
import matplotlib.pyplot as plt

class norm1:
    def __init__(self, a1, b1, c1):
        self.a1 = a1
        self.b1 = b1
        self.c1 = c1
        
    def dist_curve(self):
        plt.plot(self.c1, 1/(self.b1 * np.sqrt(2 * np.pi)) *
            np.exp( - (self.c1 - self.a1)**2 / (2 * self.b1**2) ), linewidth=4, color='y')
        plt.show()

#Vary the mean and SD to generate different plots
mean1 = 0
sd1 = 1

c = np.random.normal(mean1, sd1, 3000)
        
w1, x1, z1 = plt.hist(c, 100, density=True) #hist

hist1 = norm1(mean1, sd1, x1)
plot1 = hist1.dist_curve()
```
